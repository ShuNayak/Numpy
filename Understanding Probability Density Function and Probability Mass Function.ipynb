Probability Density Functions: Speaks of probability of given range of values occuring. In a normal distribution, the probability of landing within 1 SD of mean is 34.1%, 2 SD is 13.6% and so on. This is for continuous data.

For discrete data, probability of range of values is probability mass function.Histogram used to pictorize this.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plot

####uniform distribution means there is a flat constant probability of a value occuring
values = np.random.uniform(-10,10,100000)
#### values ranging from -10 to 10 and a 100000 of them
plot.hist(values, 50)
plot.show()


```

```python
#### visualizing the pdf
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3,3,0.001)
#### creating a list of numbers between -3 to 3 with increments of 0.001 between them
plt.plot(x,norm.pdf(x))
```

```python
#### exponential pdf
%matplotlib inline
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0,10,0.01)
plt.plot(x,expon.pdf(x))
```

```python
#### poisson probability mass function
#### ways to predict getting someother value on a future date
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import poisson

mu =500
x = np.arange(400,600,0.5)
plt.plot(x, poisson.pmf(x,mu))
```
