A library in python to make pretty graphs.

```python
# drawing a line graphs
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(-5,5,0.001)
plt.plot(x, norm.pdf(x))
plt.show()

```

```python
#drawing multiple plots on one graph
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(-10,10,1)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x,0.1,0.25))
plt.show()

#saving the plot to a file
plt.savefig('multiplePlotTest',format='jpeg')
```

```python
#adjusting the axes and including the grid lines, adding legend
y = np.arange(-6,6,1)
axes = plt.axes()
axes.set_xlim([-6,6])
axes.set_ylim([0,2.5])
axes.set_xticks([-6,-4,-2,0,2,4,6])
axes.set_yticks([0,1,2.5])
axes.grid()
plt.plot(y, norm.pdf(y),'-.')
plt.legend(['matplotlib test'], loc=4)
plt.show()
```

Pie Charts

```python
# a pie chart takes in values , colors , labels, explode proprtions and table header
values =[20,12,456,8]
colors =['b','g','y','w']
explode =[0,0,0.4,0]
labels =['hi','hello','howdy','heya']
plt.pie(values, colors= colors, labels= labels, explode=explode)
plt.title('testing pie chart in matplotlib')
plt.show()
```

```python
# a bar chart
values =[1,2,3,4]
colors= ['b','g','y','r']
plt.bar(range(0,4), values, color=colors)
plt.show()
```

```python
#scatter plot
x = np.random.randint(10, high=200,size=100)
y = np.random.randint(20,high=100, size=100)
plt.scatter(x,y)
plt.show()
```

```python
#histogram
incomes = np.random.normal(100,10,500)
plt.hist(incomes,25, color='g')
plt.show()
```
