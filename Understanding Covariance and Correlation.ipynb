Seeing if two different attributes are related to each other or not in a set of data.
1. Correlation does not imply causation , just because there is correlation between two variables does not mean one causes the other.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def find_mean(x):
    xMean = np.mean(x)
    return [xi-xMean for xi in x]

def covariance(x,y):
    n = len(x)
    return np.dot(find_mean(x),find_mean(y))/(n-1)

setOne = np.random.normal(1000, 200, 500)
setTwo = np.random.normal(1001, 300, 500)

plt.scatter(setOne, setTwo)
covariance(setOne,setTwo)
```

```python
np.corrcoef(setOne, setTwo)
# here it means that setOne has a correlation of 1 with itself , so does setTwo with itself . The rest is with each other.
```
