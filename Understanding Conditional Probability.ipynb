Probability of something occuring given that something else has occured first.
For ex: When we buy something online, the recommendations like people who bought this also bought can be thought of as the probability of a buying something when we have already bought a previous item.

P(B|A) = P(A,B)/P(A)
Here P(A,B) is the probability of both A and B occuring. 
P(B|A) is the probability of B occuring given A has already ocuured.

```python
import numpy as np
from numpy import random 
np.random.seed(10)  ## to get consistent results every time this cell is run

ages ={20:0,30:0,40:0,50:0,60:0,70:0}
purchases = {20:0,30:0,40:0,50:0,60:0,70:0}

totalPurchases =0
for _ in range(10000):
    ageRange = random.choice([20,30,40,50,60,70])
    purchaseprob = ageRange/100
    ages[ageRange]+=1
    if(random.random() < purchaseprob):
        totalPurchases+=1
        purchases[ageRange]+=1
```
